```python
import pandas as pd
```

```python
data = pd.read_csv('IBM Daily OHLC.csv')
```

```python
data
```

```python
weekday_one_hot = pd.get_dummies(data.Weekday)
data = pd.concat([data, weekday_one_hot], axis=1)
data.drop('Weekday', axis=1, inplace=True)
```

```python
data.rename({2: 'Monday', 3: 'Tuesday', 4: 'Wednesday', 5: 'Thursday', 6: 'Friday'}, axis=1, inplace=True)
```

```python
month_one_hot = pd.get_dummies(data.Month)
data = pd.concat([data, month_one_hot], axis=1)
data.drop('Month', axis=1, inplace=True)
```

```python
data.rename({1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: "July", 8: "August", 9: 'September', 10: 'October', 11: 'November', 12: 'December'}, axis=1, inplace=True)
```

```python
data
```

```python
day_one_hot = pd.get_dummies(data.Day)
data = pd.concat([data, day_one_hot], axis=1)
data.drop('Day', axis=1, inplace=True)
```

```python
data.to_csv('IBM Daily OHLC Date Encoded.csv')
```
